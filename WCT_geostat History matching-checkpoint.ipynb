```python
y = data[['REF','Sim#01','Sim#02','Sim#03','Sim#04','Sim#05','Sim#06','Sim#07','Sim#08','Sim#09','Sim#10','Sim#11','Sim#12','Sim#13'
         ,'Sim#14','Sim#15','Sim#16','Sim#17','Sim#18','Sim#19','Sim#20','Sim#21','Sim#22','Sim#23','Sim#24','Sim#25','Sim#26'
         ,'Sim#27','Sim#28','Sim#29','Sim#30']].values
x =data.iloc[:,0].values

x=x.reshape(946,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.2, random_state =42)
```

```python
model = MultiOutputRegressor(XGBRegressor(random_state=0, n_estimators =200, learning_rate=0.1))

model.fit(x_train, y_train)
```

```python
y_predict_1 = model.predict(x_train)
y_predict_2 = model.predict(x_test)
```

```python
plt.figure(figsize=(20,10), facecolor ='white')
plt.xlabel('Days'), plt.ylabel('water cut'), plt.title('Field water cut')


plt.plot(History.iloc[:,0],History.iloc[:,1], lw =5, c='b')
plt.plot(x_train,y_predict_1,'go')
plt.plot(x_test,y_predict_2,'ro')
plt.legend(['History water cut',
            'XGBoosting predicted water cut (Training)',
            'XGBoosting predicted water cut (Testing)'],loc='upper right')
```
